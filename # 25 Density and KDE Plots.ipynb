```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
# Create a quick dataframe
my_df = pd.DataFrame(randn(500, 4), columns=["Mon", "Tues", "Wed", "Thur"])
my_df
```

```python
my_df.plot(kind="kde")
```

```python
# KDE second way to Learn shift tab
my_df.plot.kde()
```

```python
# Density Plot
my_df.plot.density()
```

```python
# Othe method
my_df.plot(kind="density")
```

```python
# Title
my_df.plot(kind="density", title="My density plot")
```

```python
# change stuff
my_df.plot(kind="density", title="My density plot", alpha=.4, lw=4, grid=True)
```

```python
my_df["Tues"].plot(kind="density", title="My density plot", alpha=.4, lw=4, grid=True)
```
