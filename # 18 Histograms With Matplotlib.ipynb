```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
# Create a quick dataframe
my_df = pd.DataFrame(randn(100, 4), columns=["Mon", "Tues", "Wed", "Thur"])
my_df
```

```python
# Histagram
my_df["Wed"].hist()
```

```python
# Histagram with Smaller bins , 50, 20
my_df["Wed"].hist(bins=50)
```

```python
# Antoher way without grid bars
my_df["Wed"].plot(kind="hist")
```

```python
my_df["Wed"].hist(bins=20, grid = False, legend=True)
```
