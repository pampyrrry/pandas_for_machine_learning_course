```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
# Create a quick dataframe
my_df = pd.DataFrame(randn(500, 4), columns=["Mon", "Tues", "Wed", "Thur"])
my_df
```

```python
# Boxplots
```

```python
my_df.plot(kind="box")
```

```python
# Boxplots Size
my_df.plot(kind="box", figsize=(10,5))
```

```python
# Boxplots Color
my_df.plot(kind="box", figsize=(10,5), cmap="jet")
```

```python
# Legend - doesn't work!
```

```python
# Title
my_df.plot(kind="box", figsize=(10,5), cmap="jet", title="Boxplot")
```

```python
# Other Way
my_df.plot.box(title="My boxplots 2nd way", grid="True")
```
