```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
# Create a quick dataframe
my_df = pd.DataFrame(randn(25, 4), columns=["Mon", "Tues", "Wed", "Thur"])
my_df
```

```python
# Bar plot
my_df.abs().plot(kind="bar")
```

```python
# Bar plot stacked 
my_df.abs().plot(kind="bar", stacked=True)
```

```python
# Shading
my_df.abs().plot(kind="bar", stacked=True, alpha=0.4)
```

```python
# Title
my_df.abs().plot(kind="bar", stacked=True, alpha=0.4, title="My Bar Char")
```

```python
# Legend
my_df.abs().plot(kind="bar", stacked=True, alpha=0.4, title="My Bar Char", legend=False)
```

```python
# Seocnd WAY!
my_df.abs().plot.bar()
```

```python
# Second Way stacked
my_df.abs().plot.bar(stacked=True)
```

```python
# Second Way Shading
my_df.abs().plot.bar(stacked=True, alpha=.5)
```

```python
# Second way title
my_df.abs().plot.bar(stacked=True, alpha=.5, title="Second Way")
```

```python
# Second Way legend
my_df.abs().plot.bar(stacked=True, alpha=.5, title="Second Way", legend = False)
```

```python
# Grid lines
my_df.abs().plot.bar(stacked=True, alpha=.5, title="Second Way", grid= True)
```
