```python
import pandas as pd
import numpy as np
```

```python
# import csv file
my_df = pd.read_csv('dog_data_short.csv')
my_df
```

```python
# Add Column From List
gender = ["Male", "Female", "Male", "Male", "Female"]
my_df["Gender"] = gender
my_df
```

```python
# Add Default Values
my_df["Alive/Dead"] = [True]*len(my_df)
my_df
```

```python
# Add Nan Values
my_df["Show Dog"] = [np.nan]*len(my_df)
my_df
```

```python
# Add coulmn with .insert() - Allows Position
my_df.insert(8, "Adopted", [True]*len(my_df), True)
my_df
```

```python
# Add column with assign() - creates new df (data frame)
my_df2 = my_df.assign(Horse=[False]*len(my_df))
my_df2
```
