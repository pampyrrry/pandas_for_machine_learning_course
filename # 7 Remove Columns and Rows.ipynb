```python
import pandas as pd
import numpy as np
```

```python
# Import CSV file
my_df = pd.read_csv('dog_data_short.csv')
gender = ["Male", "Female", "Male", "Male", "Female"]
my_df["Gender"] = gender
my_df
```

```python
# Remove Column - Axis 0 Axis is column ( 0,1,2,3,4) Axis 1 is Head row- Breed, color etc
my_df.drop("Gender", axis=1) # This way doesn't delete Gender column forever. It's only doesn't show Gender column.
```

```python
# Show DataFrame Again
my_df
```

```python
# Remove column forever
my_df.drop("Gender", axis=1, inplace=True) # This way delete Gender forever - inplace arugment !
```

```python
# Show DataFrame Again
my_df
```

```python
# Remove row
my_df.drop(3, axis = 0) # This way doesn't delete 3 row forever. It's only doesn't show a row.
```

```python
my_df
```

```python
my_df.drop(3, axis = 0, inplace = True) # This way delete 3 row forever. Inplace argument.
```
