```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
# Create a quick dataframe
my_df = pd.DataFrame(randn(500, 4), columns=["Mon", "Tues", "Wed", "Thur"])
my_df
```

```python
# Hexplots
my_df.plot(kind="hexbin", x="Mon", y="Tues")
```

```python
# Hexplots size
my_df.plot(kind="hexbin", x="Mon", y="Tues", gridsize=10)
```

```python
# Hexplots color
my_df.plot(kind="hexbin", x="Mon", y="Tues", gridsize=10, cmap="Wistia")
```

```python
# Other method
my_df.plot.hexbin(x="Mon", y="Tues", gridsize=10, cmap="Wistia")
```

```python
# Title
my_df.plot.hexbin(x="Mon", y="Tues", gridsize=10, cmap="Wistia", title="My hexplot")
```

```python
# Grid
my_df.plot.hexbin(x="Mon", y="Tues", gridsize=10, cmap="Wistia", title="My hexplot", grid=True)
```
