```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
# Create a quick dataframe
my_df = pd.DataFrame(randn(25, 4), columns=["Mon", "Tues", "Wed", "Thur"])
my_df
```

```python
# Line Plot
my_df.plot(kind="line")
```

```python

my_df.plot.line()
```

```python
# Line plot absolute
my_df.abs().plot(kind="line")
```

```python
# line width
my_df.plot(kind="line", lw=3)
```

```python
# Line single
my_df["Mon"].abs().plot(kind="line")
```

```python
# LMore than one line
my_df[["Mon", "Tues"]].abs().plot(kind="line")
```

```python
# Size
my_df[["Mon", "Tues"]].abs().plot(kind="line", figsize=(10,5))
```

```python
# Tile
my_df[["Mon", "Tues"]].abs().plot(kind="line", figsize=(10,5), title="My line")
```

```python
#  Shading
my_df[["Mon", "Tues"]].abs().plot(kind="line", figsize=(10,5), title="My line", alpha=.5)
```

```python
# Legend 
my_df[["Mon", "Tues"]].abs().plot(kind="line", figsize=(10,5), title="My line", alpha=.5, legend=False)
```

```python
# Second way
my_df.plot.line()
```
