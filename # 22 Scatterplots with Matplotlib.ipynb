```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
# Create a quick dataframe
my_df = pd.DataFrame(randn(500, 4), columns=["Mon", "Tues", "Wed", "Thur"])
my_df
```

```python
# Scatteer plot 
my_df.plot(kind="scatter", x="Mon", y="Tues", c="Wed")
```

```python
# Colors: https://matplotlib.org/stable/users/explain/colors/colormaps.html
my_df.plot(kind="scatter", x="Mon", y="Tues", c="Wed", cmap="Set3")
```

```python
# By size
my_df.plot(kind="scatter", x="Mon", y="Tues", s=my_df["Wed"]*100, cmap="Set3")
```

```python
# By Szie easier to read with Aplha
my_df.plot(kind="scatter", x="Mon", y="Tues", s=my_df["Wed"]*100, alpha=.4)
```

```python
# Title Etc
my_df.plot(kind="scatter", x="Mon", y="Tues", s=my_df["Wed"]*100, alpha=.4, title="My Scatter")
```
